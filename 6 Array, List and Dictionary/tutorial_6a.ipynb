### 3
Write a program, in pseudocode, to read the name and the closing prices of several
stocks for each weekday of last week into two parallel arrays. For each stock, the
program should print the maximum and minimum price and the day it was achieved.

```python
stock_num = int(input("number of stocks to enter: "))
stock_name = []
stock_price = [[] for _ in range(stock_num)]
# write data to parallel lists
for i in range(stock_num):
    stock_name.append(input(f"item {i+1} name: "))
    for j in range(7):
        stock_price[i].append(float(input(f"item {i+1} day {j+1} price: $")))
# print max and min for each stock and day achieved
for i in range(len(stock_name)):
    min_price = min(stock_price[i])
    max_price = max(stock_price[i])
    print(f"{stock_name[i]}:\tmin: ${min_price}, day {stock_price[i].index(min_price)+1}\tmax: ${max_price}, day {stock_price[i].index(max_price)+1}")
```
