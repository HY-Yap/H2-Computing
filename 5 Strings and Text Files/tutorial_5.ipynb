### 6
Assume that a file named word.txt contains words separated by newlines. Write a code
segment that opens a file for input and prints the number of four-letter words in the file.

```python
with open('word.txt', 'r') as f:
    words = f.read().split('\n')
count = 0
for word in words:
    if len(word) == 4:
        count += 1
print(count)
```

### 6
Assume that a file named integer.txt contains integers separated by newlines. Write a
code segment that opens the file and prints the average value of the integers.

```python
with open('integer.txt', 'r') as f:
    numbers = f.read().split('\n')
for i in range(len(numbers)):
    numbers[i] = int(numbers[i])
print(sum(numbers)/len(numbers))
```
